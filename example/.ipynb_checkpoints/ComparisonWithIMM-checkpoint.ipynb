Comparing our algorithm to IMM on data from a switched system mimicing the dynamics of the linear hopper.

```python
import filterpy
from filterpy.kalman import KalmanFilter
import sys
sys.path.insert(0,'../src')
from hds import SwitchedKalman

#from dynamics.linear_hopper import Linear_Hopper, params

import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg
import os.path as path
```

### Load Data

```python
data_dir = "../data/"
data = np.loadtxt(path.join(data_dir, 'switched_meas.csv'), delimiter=',')
true = np.loadtxt(path.join(data_dir, 'switched_true.csv'), delimiter=',')
num_cont_states = 4
y = data[:,1:3]
x0 = np.array([3,1,0,0])
xtrue = true[:, np.r_[1,3,2,4]]
# x = [ pos top, pos bottom, vel top, vel bottom]
```

### Setup models

```python
p, dt = params()
p['alpha']=3
```

```python
F_A0 = np.eye(4)+dt*np.array( [ [0,0,1,0],
								[0,0,0,1],
								[-p['k']/p['mu'], p['k']/p['mu'], 0, 0],
								[p['k']/p['ml'], -p['k']/p['ml'], 0, 0]] )

F_A1 = np.eye(4)+dt*np.array( [ [0,0,1,0],
								[0,0,0,1],
								[-p['alpha']*p['k']/p['mu'], p['alpha']*p['k']/p['mu'], 0, 0],
								[p['alpha']*p['k']/p['ml'], -p['alpha']*p['k']/p['ml'], 0, 0]] )
```

```python
H = np.array([[1,0,0,0], [0,1,0,0]])
```

## IMM

### Setup

```python
filters = []
dynamics = [ F_A0, F_A1 ]
Qinv = np.diag([10,10., 1., 1])
Rinv = np.diag([1.,1.])
Q = np.diag([.1, .1, 1,1])
R = np.diag([.1,.1])
for dyn in dynamics:
	filt = KalmanFilter(dim_x = 4, dim_z = 2)
	filt.F = dyn
	filt.H = H
	filt.x = x0
	filt.P *= 1e-2
	#filt.R *= .01
	#q = np.array([[ .05, .125], [.125, 1/3]])
	#filt.Q = filterpy.common.Q_discrete_white_noise(dim=4, dt=dt, var=1e-5)
	#filt.Q = scipy.linalg.block_diag(q,q)*1e-2
    
	filt.R = R
	filt.Q = Q
	filters.append(filt)

k = len(dynamics)
mu = np.ones(k)/k
prob = .01
trans = np.ones((k,k))*prob
np.fill_diagonal(trans, 1-(k-1)*prob)

imm = filterpy.kalman.IMMEstimator(filters,mu,trans)
```

### Run

```python
imm_xhat = []
imm_what = []

for meas in data:
	imm.predict()
	imm.update(meas[1:])
	imm_xhat.append(imm.x.copy())
	imm_what.append(imm.mu.copy())

imm_xhat = np.vstack(imm_xhat)
imm_what = np.vstack(imm_what)
```

```python
fig,ax= plt.subplots(3,1, figsize=(6,8), sharex=True)
ax[0].plot(imm_xhat[:,:2]-xtrue[:,:2])
ax[1].plot(imm_xhat[:,2:] - xtrue[:,2:])
ax[2].plot(imm_what);
#ax[2].plot(true[:,-1])
```
